```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
path_data = '../data_asteroid_regression.csv'

# data = np.genfromtxt('data_asteroid_regression.csv', delimiter=',')
# d, y = data.T

df_data = pd.read_csv(path_data, header=None, names = ['d','m']) 
df_data = df_data.sort_values(by='d')
df_data
```

```python
fig, ax = plt.subplots(figsize=(12,9))
ax.plot(df_data.d, df_data.m, 'bx')
ax.grid()
```

```python
A = np.array([df_data.d.values**1, df_data.d.values**2, df_data.d.values**3]).T
# np.column_stack([df_data.d.values, df_data.d.values**2, df_data.d.values**3])
A
```

```python
y = np.array([df_data.m.values]).T
y
```

```python
x_min = np.matmul( 
    np.matmul( 
        np.linalg.inv( np.matmul(A.T, A) ),
        A.T
    ), y
)
x_min
```

```python
fig, ax = plt.subplots(figsize=(12,9))
ax.plot(df_data.d, df_data.m, 'bx', label='data')
ax.plot(df_data.d, np.matmul(A, x_min), 'r--', label='l2 optimal')
ax.grid()
ax.legend()
```

## L1 Norm Gradient Descent

```python
x0 = np.array([[0],[0],[0]])
step_size = 1e-4
R = np.matmul(A, x0) - y
grad = np.matmul(A.T, np.sign(R))
grad
```

```python
x_best = None
error_best = np.inf
x_next = x0
for i in range(int(1e4)):
    # Calculate residual vector
    R = np.matmul(A, x_next) - y

    # Calculate L1 Norm for residual
    error_l1 = np.sum(abs(R))

    # Record if the current L1 Norm Residual is better than before
    if error_l1 < error_best:
        error_best = error_l1
        x_best = x_next

    # Compute gradient by the formula derived
    # Note: no explicit subgradient need to be picked because 
    # np.sign(0) returns 0, which is in the set of sub-differential already
    grad = np.matmul(A.T, np.sign(R))

    # Descend along the gradient
    x_next = x_next - step_size * grad

print('optimal fit:\n', x_best)
print('best fit error:', error_best)
```

```python
fig, ax = plt.subplots(figsize=(12,9))
ax.plot(df_data.d, df_data.m, 'bx', label='data')
ax.plot(df_data.d, A@x_min, 'r--', label='l2 optimal')
ax.plot(df_data.d, A@x_best, 'g--', label='l1 optimal')
ax.grid()
ax.legend()
```
